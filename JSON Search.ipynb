```python
import os
import json
import pandas as pd

main_path = r"E:\WataLibrary\WataUtils\chat_vod_download"

all_data = []

for file in os.listdir(main_path):
    if '.json' in file:
        path = os.path.join(main_path,file)
        with open(path, 'r',encoding="utf8") as file:
            data = json.load(file)
        all_data.append(data)

def seconds_to_hhmmss(total_seconds):
    minutes, seconds = divmod(total_seconds, 60)
    hours, minutes = divmod(minutes, 60)
    return f"{int(hours):02d}:{int(minutes):02d}:{int(seconds):02d}"
```

```python
search_term = ""

for data in all_data:
    print("-------------------------------------------------------------[{}]".format(data['video']['id']))
    for comment in data['comments']:
        message = comment['message']['body'].lower()
        if not pd.isna(comment['commenter']):
            chatter = comment['commenter']['display_name']
        else:
            chatter = ""
        time = seconds_to_hhmmss(comment['content_offset_seconds'])
        if search_term in message:
            print(f"{time}:00 {message} | {chatter}")
```
