```python
import re
import os
import pyperclip
import pyautogui
import time
import winsound

from pycaw.pycaw import AudioUtilities, ISimpleAudioVolume
sessions = AudioUtilities.GetAllSessions()
python_process = os.path.basename(os.sys.executable)
for session in sessions:
    if session.Process and session.Process.name() == python_process:
        volume = session._ctl.QueryInterface(ISimpleAudioVolume)
        volume.SetMasterVolume(0.1, None)

def task_complete_alarm():
    winsound.PlaySound(r".\resources\tuturu.wav",winsound.SND_ASYNC)
```

```python
time.sleep(2)
strX, strY = pyautogui.position()
task_complete_alarm()
```

```python
time.sleep(2)
dateX, dateY = pyautogui.position()
task_complete_alarm()
```

```python
time.sleep(2)
resetX, resetY = pyautogui.position()
task_complete_alarm()
```

```python
markers = open(r'.\input_files\markers.txt', 'r').readlines()
```

```python
for marker in markers:
    if re.compile(r'[0-9]-').search(marker):
        print(marker.split(" ")[1].split("-")[0])
```

```python
last_marker = ""
for marker in markers:
    if marker.split(" ")[0] == last_marker:
        print(marker)
    last_marker = marker.split(" ")[0]
```

```python
pyautogui.click(resetX,resetY)
time.sleep(3)
for marker in markers:
    pyautogui.hotkey('ctrl', 'v')
    time.sleep(1)
    pyautogui.hotkey('3')

time.sleep(5)
pyautogui.click(resetX,resetY)
pyautogui.hotkey('home')
pyautogui.hotkey('s')
time.sleep(5)
```

```python
pyautogui.click(resetX,resetY)
pyautogui.hotkey('home')

time.sleep(3)
for marker in markers:
    pyautogui.hotkey('right') 
    dash_flag = False
    if re.compile(r'[0-9]-').search(marker):
        dash_flag = True
        title = '-'.join(marker.split("-")[1:])[:-1]
        day = int(marker.split("-")[0].split(" ")[1])
        if 4 <= day <= 20 or 24 <= day <= 30:
            suffix = "th"
        else:
            suffix = ["st", "nd", "rd"][day % 10 - 1]
    else:
        title = marker[9:]
    date_str = "March {0}{1} 2026".format(day,suffix) #CHANGE
    pyperclip.copy(title)
    pyautogui.tripleClick(strX,strY)
    pyautogui.hotkey('ctrl', 'v')
    if not dash_flag:
        pyautogui.hotkey('backspace')
    time.sleep(1)
    pyperclip.copy(date_str)
    pyautogui.tripleClick(dateX,dateY)
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.click(resetX,resetY)
    pyautogui.hotkey('3')
    time.sleep(3)
task_complete_alarm()
```
